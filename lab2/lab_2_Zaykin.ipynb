Теперь побудуємо бінарний МНК-класифікатор та виведо результати на графіку

```python
import numpy as np
import matplotlib.pyplot as plt

x_train = [x + [1] for x in x_train]  # Додаємо зміщення для кожного прикладу
x_train = np.array(x_train)
y_train = np.array(y_train)

pt = np.sum([x * y for x, y in zip(x_train, y_train)], axis=0)  # Обчислення підсумку
xxt = np.sum([np.outer(x, x) for x in x_train], axis=0)  # Обчислення підсумку зовнішнього добутку
w = np.dot(pt, np.linalg.inv(xxt))  # Обчислення вагових коефіцієнтів
print(f"Вектор ваг: {w}")

line_x = list(range(max(x_train[:, 0])))    # формування координат для лінії розділення
line_y = [-x*w[0]/w[1] - w[2]/w[1] for x in line_x]

x_0 = x_train[y_train == 1]                 # формування точок для класу 1
x_1 = x_train[y_train == -1]                # і класу -1

plt.scatter(x_0[:, 0], x_0[:, 1], color='red')  # відображення точок класу 1 червоним кольором
plt.scatter(x_1[:, 0], x_1[:, 1], color='blue')  # відображення точок класу -1 синім кольором
plt.plot(line_x, line_y, color='green')  # відображення лінії розділення зеленим кольором

plt.xlim([0, 45])
plt.ylim([0, 75])
plt.ylabel("довжина")
plt.xlabel("ширина")
plt.grid(True)
plt.show()
```

1. Що таке відступ (margin) і яка його роль в побудові моделі класифікації?

	Відступ - це міра правильності прогнозу для кожного об'єкта. Якщо відступ $< 0$, то це означає що передбачення помилкове і воно потребує коригування, якщо $> 0$ - передбачення правильне і ніяких змін модель не потребує.

2. Яка роль функції втрат і чому не можна знайти коефіцієнти лінійного класифікатора шляхом безпосереднього розв’язаня задачі оптимізації через знаходження похідної?

	Для знаходження МНК-оцінки классифікатора треба взяти похідну функції втрат і прирівняти її до нуля. Через неї знаходится такий вектор ваг $w$, який досягає мінімума помилок на графіку $a(x,w)$. Але така функція не є диференційованою, тому що має різкий розрив. Щоб знайти похідну підбирається формула суми найменших квадратів (функціонал якості), яка може бути використана для підрахунку сумарного штрафу по всій вибірці і може бути диференційована.

3. Чим приницпово розрізняється алгоритм, розглянутий у цій роботі від алгоритму, розглянутого у лаб. роб 1?

	МНК розв’язує задачу класифікації як задачу оптимізації, тобто шукає мінімуми оба максимуми фукнції. Алгоритм Розенблатта - це евристичне правило без оптимізації: він просто перебирає об’єкти і коригує вагу лише тоді, коли трапляється помилка.
