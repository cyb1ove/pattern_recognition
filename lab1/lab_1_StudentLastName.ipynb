Тепер реалізуємо алгоритм бінарного класифікатора мовою Python.

```python
import numpy as np
import matplotlib.pyplot as plt

# Реалізація перцептрона Розенблатта для 2D-ознак
x_train = np.array(x_train)
y_train = np.array(y_train)
n_train = len(x_train)
N = 200               # максимальна кількість ітерацій (епох)
L = 0.1               # крок навчання

w = np.zeros(2)       # ваги для x1, x2
b = 0.0               # зсув (bias)

for n in range(N):
    errors = 0
    for i in range(n_train):
        margin = y_train[i] * (np.dot(w, x_train[i]) + b)

        # Якщо точка помилково класифікована або лежить на межі,
        # виконуємо стандартне оновлення перцептрона
        if margin <= 0:
            w = w + L * y_train[i] * x_train[i]
            b = b + L * y_train[i]
            errors += 1

    if errors == 0:
        break

Q = sum(
    1 for i in range(n_train)
    if y_train[i] * (np.dot(w, x_train[i]) + b) <= 0
)

print(f"Кількість помилок: {Q}; Вектор ваг: {np.round(w, 3).tolist()}; bias: {round(b, 3)}")

# Побудова роздільної прямої: w0*x + w1*y + b = 0
line_x = np.linspace(x_train[:, 0].min() - 1, x_train[:, 0].max() + 1, 200)
if abs(w[1]) > 1e-12:
    line_y = -(w[0] * line_x + b) / w[1]
else:
    # Вертикальна пряма, якщо w1 близький до нуля
    line_y = np.full_like(line_x, np.nan)
    x_vertical = -b / w[0] if abs(w[0]) > 1e-12 else 0

x_0 = x_train[y_train == 1]
x_1 = x_train[y_train == -1]

plt.scatter(x_0[:, 0], x_0[:, 1], color='red', label='class +1')
plt.scatter(x_1[:, 0], x_1[:, 1], color='blue', label='class -1')

if abs(w[1]) > 1e-12:
    plt.plot(line_x, line_y, color='green', label='decision boundary')
else:
    plt.axvline(x=x_vertical, color='green', label='decision boundary')

plt.xlim([0, 45])
plt.ylim([0, 75])
plt.ylabel("довжина")
plt.xlabel("ширина")
plt.legend()
plt.grid(True)
plt.show()
```

**Висновок**: Навчальна вибірка не є лінійно роздільною для обраної моделі лінійного бінарного класифікатора. На графіку видно, що точки класів +1 і -1 не можуть бути повністю розділені однією прямою. Отже, алгоритм не досягає нульової кількості помилок (Q = 0), що підтверджує обмеження лінійного класифікатора для даного набору даних.

1. Які основні принципи лежать в основі розпізнавання образів у контексті машинного навчання?

	Розпізнавання образів базується на тому, що кожен об’єкт описується вектором ознак $x = (x_1, x_2, \ldots, x_n)^T$, а алгоритм на основі навчальної вибірки пар $(x_i, y_i)$ шукає межу розділення між класами у просторі ознак.

2. Що таке навчальна вибірка і чому вона важлива для процесу розпізнавання образів?

	Навчальна вибірка це пари $(x_i, y_i)$, де $x$ - матриця ознак об'єктів, $y$ - клас об'єкта. В алгоритмах навчання з вчителем, які використовуються для задач класифікації і задач регресії необхідна навчальна вибірка, щоб вивчити закономірність відомих пар ознак і класів.

3. Зформулювати постановку задачі бінарної класифікації.

	Задача бінарної класифікації полягає в тому, щоб побудувати межу в $n$-вимірному просторі ознак, з якою усі об'єкти одного класу мають знаходитися з одного боку межі, а всі об'єкти іншого класу — з іншого боку. Для лінійних моделей це буде гіперплощина, для не	лінійних - більш складні форми межі.

4. Що таке рівняння гіперплощини і як воно використовується для розв’язання задачі бінарної класифікації?

	Рівняння гіперплощини — це лінійне рівняння виду $\omega^T x = 0$, яке описує межу розділення між класами у просторі ознак. Потрібно підібрати такий вектор $\omega$, щоб гіперплощина якнайкраще розділяла об’єкти класів із навчальної вибірки.
